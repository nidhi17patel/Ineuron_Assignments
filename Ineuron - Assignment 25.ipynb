1) . What is the difference between enclosing a list comprehension in square brackets and
parentheses?

When we enclose a list comprehension in square brackets, you create a new list containing the values generated by the list comprehension.
When we enclose a list comprehension in parentheses, you are actually creating a generator expression.

2) What is the relationship between generators and iterators?

Generators are a type of iterator in Python. Generators are a convenient way to create iterators using a more concise syntax. While both generators and iterators allow you to iterate over sequences of data, generators are more memory-efficient because they generate values on-the-fly rather than creating an entire sequence in memory like a list.
